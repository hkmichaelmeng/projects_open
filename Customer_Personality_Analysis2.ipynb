## Data processing

As a unsupervised clustering project, I want to get features that are really related with the final results. According domain knowledge of customer behavior, features including age, income, historical purchasing, behaviors against promotions and complains may directly influence the customers future buying opportunity.
There are several columns to deliver the same feature details, but we need to combine them for the sake of clustering models' performance.

```python
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'], dayfirst=True,format = '%d-%m-%Y')

data['Dt_Customer'].agg(['min','max'])
```

```python
import datetime
from datetime import date
last_date = date(2014,12,31) #Simulating data analysis time. 
data['Seniority'] = pd.to_numeric(data['Dt_Customer'].dt.date.apply(lambda x: (last_date - x)).dt.days, downcast='integer')/30
data['Seniority']
```

```python
data['Age']=2014-data['Year_Birth']  
```

```python
data['Age'].dtype
```

```python
import numpy as np

data['Spending']=data['MntWines']+data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntSweetProducts']+data['MntGoldProds']
data['Spending']
```

```python
data.dropna(subset=['Income'],inplace=True)
data.shape
```

```python
data[data['Income']>300000].count()
```

```python
data=data[data['Income']<=300000]
data.shape
```

```python
data['kids']=data['Kidhome']+data['Teenhome']
data['purchase_total']=data['NumWebPurchases']+data['NumCatalogPurchases']
+data['NumStorePurchases']
```

```python
data.columns
```

```python
data_temp=data[['Income','Seniority','Spending','Age','kids', 'purchase_total','NumDealsPurchases']]
data_temp.describe().T
```

## Clustering
Stars: Old customers with high income and high spending nature.
Need Attention: New customers with below-average income and low spending nature. 
High Potential: New customers with high income and high spending nature.
Leaky Bucket: Old customers with below-average income and a low spending nature.

As data summary above, I will standardise income, spending, NumDealsPurchases and purchase_total features because it is more likely standardised distribution. And normalise Seniority and Age, while pass kids number directly to the models.

```python
from sklearn.preprocessing import StandardScaler, normalize

scaler=StandardScaler()
data_temp2=data[['Income','NumDealsPurchases','Spending','purchase_total']]
X_std=scaler.fit_transform(data_temp2)
```

```python
X_std
```

```python
X=data['kids'].values
```

```python
X=np.reshape(X,(2215,1))
```

```python
X.ndim
```

```python
X_std2=normalize(data[['Seniority','Age']], norm='l2')
type(X_std2)
```

```python
X_Total=np.concatenate([X_std, X, X_std2], axis=1)
X_Total.shape
```

```python
from sklearn.mixture import GaussianMixture
gmm=GaussianMixture(n_components=4, covariance_type='spherical',max_iter=2000, random_state=1).fit(X_Total)
labels = gmm.predict(X_Total)  #only pass 3 features to GaussianMixture.Fit data and predict data are same
labels
```

```python
data_temp['Cluster'] = labels
data_temp.groupby('Cluster').mean()
```

```python
data_temp.groupby('Cluster').describe().T
```

```python
data_temp['Cluster']=data_temp['Cluster'].replace({0:'Potential',1:'Discount_buyer',2:'VIP',3:'LowValue'})
data_temp
```
